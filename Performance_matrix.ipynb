**Performance Matrix**

```python
from sklearn.metrics import *
```

```python
actual =    [1,0,1,1,0,1,0,0,1,0,1]
predicted = [1,0,1,0,0,1,1,1,1,0,1]
```

Confusion Matrix

```python
confu_mat = confusion_matrix(actual, predicted)
print(confu_mat)
```

True Positives (TP) = 5 → Correctly predicted positive cases

True Negatives (TN) = 3 → Correctly predicted negative cases

False Positives (FP) = 2 → Incorrectly predicted positive cases

False Negatives (FN) = 1 → Positive cases that the model failed to detect

```python
TP = confu_mat[1,1]
TN = confu_mat[0,0]
FP = confu_mat[0,1]
FN = confu_mat[1,0]
```

Accuracy

```python
accu = accuracy_score(actual,predicted)
print(accu)
```


Accuracy (0.727): The model correctly predicts 72.7% of the total samples.

```python
accuracy_formula = (TP + TN)/(TP+FP+TN+FN)
print(accuracy_formula)
```

Precision

```python
precision_s = precision_score(actual,predicted)
print(precision_s)
```

Precision (0.714): Out of all predicted positives, 71.4% are actually positive.

```python
precision_s_formula = (TP)/(TP+FP)
print(precision_s_formula)
```

Recall

```python
recall = recall_score(actual,predicted)
print(recall)
```

Recall (0.833): The model correctly identifies 83.3% of the actual positive cases.

```python
recall_formula = (TP) / (TP + FN)
print(recall_formula)
```

Fbeta Score


```python
f1score = f1_score(actual,predicted)
print(f1score)
```

F1 Score (0.769): The model has a balanced performance between precision and recall.

```python
f1score_formula = (2 *precision_s * recall)/(precision_s+recall)
print(f1score_formula)
```

```python
f2score = fbeta_score(actual,predicted,beta=2)
print(f2score)
```

F2 Score (~0.806): Gives more importance to Recall, meaning the model focuses more on detecting positive cases.

```python
f05score = fbeta_score(actual,predicted,beta=0.5)
print(f05score)
```

F0.5 Score (~0.735): Gives more importance to Precision, meaning the model focuses more on reducing false positives.

```python
specificity = TN / (TN+FP)
print(specificity)
```

```python
False_Positive_Rate = FP / (FP+TN)
print(False_Positive_Rate)
```

```python
False_Negative_Rate = FN / (FN+TP)
print(False_Negative_Rate)
```
